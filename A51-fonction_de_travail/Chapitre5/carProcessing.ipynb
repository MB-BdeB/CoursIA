4. Détection des valeurs Aberrantes : Score Z

```python
from scipy.stats import zscore

# Colonnes numériques à analyser
numerical_cols = ['prix', 'cylindree', 'puissance', 'poids', 'conso']

# Dictionnaire pour stocker les valeurs aberrantes
zscore_outliers_summary = {}

# Détection des valeurs aberrantes avec la méthode Z-score
for col in numerical_cols:
    z_scores = zscore(df1[col])
    outliers = df1[(z_scores > 2) | (z_scores < -2)]
    zscore_outliers_summary[col] = outliers[['type', col]]

# Affichage des résultats
for col, outliers in zscore_outliers_summary.items():
    print(f"\nValeurs aberrantes pour '{col}' (méthode Z-score):")
    print(outliers)
```

```python
# Définir les seuils pour identifier les outliers
threshold1 = 2 # Seuil 1 pour détecter les outliers (écart-type * 2)
threshold2 = 3 # Seuil 2 pour détecter les outliers (écart-type * 3)
# Créer des copies du DataFrame original pour travailler sur les outliers sans modifier df directement
dfOutliers1 = df.copy()
dfOutliers2 = df.copy()
# Afficher la dimension initiale du DataFrame df
print("Dimension de la base de df:",df.shape)
# Calculer les bornes supérieures et inférieures pour le seuil 1
upper1 = dfOutliers1.prix.mean() + threshold1*dfOutliers1.prix.std() # Limite supérieure : moyenne + seuil * écart-type
lower1 = dfOutliers1.prix.mean() - threshold1*dfOutliers1.prix.std() # Limite inférieure : moyenne - seuil * écart-type
# Filtrer les lignes du DataFrame dfOutliers1 pour supprimer les valeurs en dehors des bornes (outliers)
dfOutliers1= dfOutliers1[(dfOutliers1.prix<upper1) & (dfOutliers1.prix>lower1)]
# Afficher la dimension après suppression des outliers pour dfOutliers1
print("Dimension de la base de dfOutliers1:",dfOutliers1.shape)
# Calculer les bornes supérieures et inférieures pour le seuil 2
upper2 = dfOutliers2.prix.mean() + threshold2 * dfOutliers2.prix.std() # Limite supérieure : moyenne + seuil * écart-type
lower2 = dfOutliers2.prix.mean() - threshold2 * dfOutliers2.prix.std() # Limite inférieure : moyenne - seuil * écart-type
# Filtrer les lignes du DataFrame dfOutliers2 pour supprimer les valeurs en dehors des bornes (outliers)
dfOutliers2 = dfOutliers2[(dfOutliers2.prix < upper2) & (dfOutliers2.prix > lower2)]
# Afficher la dimension après suppression des outliers pour dfOutliers2
print("Dimension de la base de dfOutliers2:", dfOutliers2.shape)
```

3. Standardisez la variable x et stocker le résultat dans x_standard

```python
# Appliquer la normalisation standard

standard_scaler = StandardScaler()
x_standard = standard_scaler.fit_transform(x)
print(np.mean(x))
print(np.mean(x_standard))
print(np.std(x_standard))

# Convertir le tableau x_minmax = scaler.fit_transform(x) en DataFrame pour une meilleure lisibilité
x_standard_df = pd.DataFrame(x_standard, columns=x.columns)
display(x_standard_df)
```

4. Repésentez l’histogramme des variables x, x_minmax et x_standard

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.hist(x, bins=10, color='skyblue', edgecolor='black')
plt.title("Min-Max Normalized 'cylindree'")
plt.xlabel("Original Value")
plt.ylabel("Frequency")

plt.subplot(1, 3, 2)
plt.hist(x_minmax, bins=10, color='skyblue', edgecolor='black')
plt.title("Min-Max Normalized 'cylindree'")
plt.xlabel("Normalized Value")
plt.ylabel("Frequency")

plt.subplot(1, 3, 3)
plt.hist(x_standard, bins=10, color='salmon', edgecolor='black')
plt.title("Z-score Standardized 'cylindree'")
plt.xlabel("Standardized Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

```

- La normalisation réduit l’échelle des données entre 0 et 1, mais conserve la forme de la distribution.
    - La moyenne n’est pas 0.5 car la distribution des cylindrées est asymétrique (il y a des voitures avec des cylindrées très élevées qui tirent la moyenne vers le haut).

- La standardisation Z-score transforme les données pour qu’elles aient :
    - une moyenne de 0
    - un écart-type de 1
    - Cela permet de centrer les données et de les rendre comparables même si elles ont des unités ou des échelles différentes.
    - Les valeurs très éloignées de la moyenne (en nombre d’écarts-types) sont facilement identifiables comme valeurs aberrantes (outliers).

|Aspect	| Min-Max Normalisation	| Z-score Standardisation
|:---------------|:--------------|:---------------|
|Échelle	|Toutes les valeurs sont entre 0 et 1	|Moyenne = 0, Écart-type = 1
|Forme de la distribution	|La forme est conservée, mais l’échelle est compressée	|La forme est centrée, les valeurs extrêmes sont plus visibles
|Sensibilité aux outliers	| Sensible aux valeurs extrêmes (elles tirent les autres vers le bas)	| Moins sensible, car les extrêmes sont exprimés en nombre d’écarts-types
|Utilisation typique	| Réseaux de neurones, distances (KNN, clustering)	| Régression, PCA, modèles linéaires

Utilise Min-Max si tu veux réduire toutes les valeurs dans une plage fixe (utile pour les algorithmes sensibles à l’échelle).
Utilise Z-score si tu veux comparer les écarts par rapport à la moyenne ou détecter des valeurs aberrantes.
