Q2. Face à un problème d’apprentissage machine, comment pouvons-nous anticiper s’il
s’agira d’utiliser une technique d’apprentissage supervisée ou une technique d’apprentissage
non-supervisée ?

Pour anticiper si une technique d’apprentissage supervisée ou non-supervisée sera utilisée, il faut analyser les données disponibles et l’objectif de l'analyse.  
- Si les données sont étiquetées, une technique d’apprentissage supervisée est généralement appropriée.  
- Si les données ne sont pas étiquetées et que l’objectif est de découvrir des structures ou des relations dans les données, l’apprentissage non-supervisée sera plus adaptée.

Q4. Qu’est qu’une modalité ?

- Une modalité est une valeur ou une catégorie possible d'une variable. Par exemple, pour une variable "Genre", les modalités peuvent être "Masculin" et "Féminin".

Q5. Le coefficient d’asymétrie d’une distribution est nul. Que peut-on dire de la distribution ?

- Si le coefficient d’asymétrie d’une distribution est nul, cela signifie que la distribution est symétrique par rapport à sa moyenne. Les valeurs sont réparties de manière égale de chaque côté de la moyenne.

# Partie II : Description des données

1. Charger les données

```python
import pandas as pd

df = pd.read_excel("C:\\Users\\bouma12\\VSCode_WorkSpace\\Fonction_travail\\examen_Fonction_Travail\\RecolteVins.xlsx")
print("Premières lignes du DataFrame:")
print(df.head())
```

2. Déterminez le nombre d’individus et de variables et identifiez le type des variables

```python
num_individuals = df.shape[0]
num_variables = df.shape[1]

# Types of variables
variable_types = df.dtypes

print(f"Nombre des individus: {num_individuals}")
print(f"Nombre des variables: {num_variables}")

print("Types des variables:")
print(variable_types)
```

3. Calculez la moyenne et l’écart-type de la variable Intensity

```python
moyenne_intensity = df['Intensity'].mean()
sigma_intensity = df['Intensity'].std()

print(f"Moyenne de la variable Intensity: {moyenne_intensity:.2f}")
print(f"Écart-type de la variable Intensity: {sigma_intensity:.2f}")
```

4. Déterminez les modalités de la variable Region

```python
modalites_region = df['Region'].unique()
print("Modalités de la variable Region:")
print(modalites_region)
```

5. Calculez l’effectif de chaque modalité de Region

```python
effectif_region = df['Region'].value_counts()
print("Effectif de chaque modalité de Region:")
print(effectif_region)
```

6. Visualiser le diagramme circulaire de la variable Region

```python
effectif_region.plot.pie(
    autopct='%1.1f%%', 
    figsize=(4, 4), 
    title="Répartition des modalités de la variable Region"
)
```

7. Déterminez les caractéristique de tendence centrale de la varible Spice

```python
moyenne_spice = df['Spice'].mean()
print(f"La moyenne de Spice est {moyenne_spice:.2f} ans")

mediane_spice = df['Spice'].median()
print(f"La médiane de Spice est {mediane_spice:.2f} ans")

quartile_1_spice = df['Spice'].quantile(0.25)
print(f"Le premier quartile d'âge est {quartile_1_spice:.2f} ans")    
quartile_2_spice = df["Spice"].quantile(0.50)
print(f"Le deuxième quartile d'âge est {quartile_2_spice:.2f} ans")
quartile_3_spice = df["Spice"].quantile(0.75)
print(f"Le troisième quartile d'âge est {quartile_3_spice:.2f} ans")
```

8. Représentez l’histogramme de la variable Spice

```python
import matplotlib.pyplot as plt

# Histogram of the variable Spice
df['Spice'].plot.hist(
    bins=10, 
    alpha=0.7, 
    color='blue', 
    title="Histogramme de la variable Spice"
)
plt.xlabel("Spice")
plt.ylabel("Fréquence")
plt.show()
```

9. Calculez le coefficient d’assymétrie de la variable Spice

```python
coefficient_asymetrie = df['Spice'].skew(skipna=True)
print(f"Le coefficient d'asymétrie de la variable 'Spice' est {coefficient_asymetrie:.2f}")

# Interprétation
if coefficient_asymetrie > 0:
    print("La distribution est asymétrique à droite.")
elif coefficient_asymetrie < 0:
    print("La distribution est asymétrique à gauche.")
else:
    print("La distribution est symétrique.")
```

10. Interprétez les résultats

La distribution est asymétrique à gauche.
