Déterminez les caractéristiques de tendances centrales de la variable
age : la moyenne, la médiane, les quartiles Q1, Q2, Q3.

```python
moyenne_age = df['age'].mean()
print(f"La moyenne d'âge est {moyenne_age:.2f} ans")
mediane_age = df['age'].median()
print(f"La médiane d'âge est {mediane_age:.2f} ans")
quartile_1_age = df['age'].quantile(0.25)
print(f"Le premier quartile d'âge est {quartile_1_age:.2f} ans")    
quartile_2_age = df["age"].quantile(0.50)
print(f"Le deuxième quartile d'âge est {quartile_2_age:.2f} ans")
quartile_3_age = df["age"].quantile(0.75)
print(f"Le troisième quartile d'âge est {quartile_3_age:.2f} ans")
```

Caractéristiques de dispersion

```python
etendue_age = df['age'].max() - df['age'].min()
print(f"L'étendue d'âge est {etendue_age:.2f} ans")
variance_age = df['age'].var()
print(f"La variance d'âge est {variance_age:.2f} ans²")
sigmame_age = df['age'].std()
print(f"L'écart-type d'âge est {sigmame_age:.2f} ans")
```

Calculez le coefficient d'asymétrie et interpréter le résultat

```python


# Calcul du coefficient d'asymétrie
coefficient_asymetrie = df['age'].skew(skipna=True)
print(f"Le coefficient d'asymétrie de la variable 'age' est {coefficient_asymetrie:.2f}")

# Interprétation
if coefficient_asymetrie > 0:
    print("La distribution est asymétrique à droite.")
elif coefficient_asymetrie < 0:
    print("La distribution est asymétrique à gauche.")
else:
    print("La distribution est symétrique.")
```

Calculez le coefficient d'aplatissement et interpréter le résultat

```python
# Calcul du coefficient d'aplatissement
coefficient_aplatissement = df['age'].kurt(skipna=True)
print(f"Le coefficient d'aplatissement de la variable 'age' est {coefficient_aplatissement:.2f}")

# Interprétation
if coefficient_aplatissement > 0:
    print("La distribution est plus pointue que la distribution normale.")
elif coefficient_aplatissement < 0:
    print("La distribution est plus aplatie que la distribution normale.")
else:
    print("La distribution a un aplatissement similaire à celui de la distribution normale.")
```
