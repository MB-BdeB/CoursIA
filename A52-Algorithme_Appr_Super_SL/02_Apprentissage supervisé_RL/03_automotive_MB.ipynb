```python
df_1 = df.drop(columns=['normalized-losses'])
df_1.columns
```

```python
# compter les nombres null ou na dans chaque colonne

# print(df_1.isnull().sum())
# Afficher uniquement les colonnes où il y a des valeurs manquantes (> 0)
print(df_1.isna().sum()[df_1.isna().sum() > 0])# Afficher uniquement les colonnes où il y a des valeurs manquantes (> 0)
```

```python
df_2 = df_1.dropna()
print(df_2.shape)
print(df_2.isna().sum()[df_2.isna().sum() > 0])# Afficher uniquement les colonnes où il y a des valeurs manquantes (> 0)
```

```python
cols = ['symboling', 'fuel-type', 'aspiration','num-of-doors', 'drive-wheels', 
        'engine-location', 'length', 'width', 'height', 'curb-weight', 'engine-type',
        'num-of-cylinders', 'fuel-system', 'bore', 'stroke','compression-ratio', 'city-mpg']
df_3 = df_2.drop(columns=cols)
df_3.info()
```

```python
df_3 = pd.get_dummies(df_3, columns=['make', 'body-style'])
df_3.info()
df_3.head()
```

```python
import matplotlib.pyplot as plt

for col in df_3.columns:
    plt.figure(figsize=(6, 4))
    plt.hist(df_3[col], bins=20, edgecolor='black')
    plt.title(f'Histogramme de {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence')
    plt.show()
```

```python
import matplotlib.pyplot as plt

predictors = df_3.columns.drop('price')
n = len(predictors)
fig, axes = plt.subplots(nrows=(n // 2) + (n % 2 > 0), ncols=2, figsize=(12, 4 * ((n // 2) + (n % 2 > 0)))
)

for i, col in enumerate(predictors):
    ax = axes.flat[i]
    ax.scatter(df_3[col], df_2['price'], alpha=0.6)
    ax.set_xlabel(col)
    ax.set_ylabel('price')
    ax.set_title(f'{col} vs price')

# Masquer les sous-graphiques vides
for j in range(i + 1, len(axes.flat)):
    axes.flat[j].set_visible(False)

plt.tight_layout()
plt.show()
```

Pour le nuage de points `city-mpg vs price` :

- **city-mpg** représente la consommation en ville (miles par gallon).
- Si tu observes une tendance descendante dans le graphique, cela signifie que plus la consommation est élevée (plus de miles par gallon), moins le prix de la voiture est élevé.
- Cela s’explique souvent car les voitures économiques (faible consommation) sont moins chères, tandis que les voitures puissantes (consommation élevée) sont plus chères.

**Conclusion :**  
Il existe généralement une **corrélation négative** entre `city-mpg` et `price` :  
Quand `city-mpg` augmente, le prix diminue.  
Ce descripteur peut donc être utile pour prédire le prix, mais il indique surtout une relation inverse.

```python
from sklearn.linear_model import LinearRegression

# Initialiser le modèle
model = LinearRegression()

# Définir les variables explicatives (X) et la variable cible (y)
columns = df_3.columns[df_3.columns != 'price']
X = df_3[columns]
y = df_3['price']
```

```python
from sklearn.model_selection import train_test_split

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
print("Ensembles d'entraînement et de test créés avec succès.\n",
      "X_train",X_train.shape, 
      "\nX_test", X_test.shape, 
      "\ny_train", y_train.shape, 
      "\ny_test", y_test.shape)
```

```python

model.fit(X_train, y_train)
```

```python
import numpy as np

print("Coefficients:", model.coef_)
print("Coefficients arrondis:", np.round(model.coef_, 2))
print("Intercept:", round(model.intercept_, 2))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2:", r2_score(y_test, y_pred))
```

### Affichage des 10 premières prédictions et valeurs réelles
L'échantillon de test (`X_test`, `y_test`) est une partie des données non utilisée pour l'apprentissage du modèle. Il permet de vérifier la capacité du modèle à généraliser sur des données inconnues.
Ci-dessous, on compare les 10 premières valeurs prédites par le modèle avec les 10 vraies valeurs du prix pour évaluer la qualité des prédictions.

```python
# Afficher les 10 premières prédictions et les 10 vraies valeurs du test
print("Premières prédictions :", np.round(y_pred[:10], 0))
print("Valeurs       réelles :", np.round(y_test[:10].values, 0))
```
