```python
#imc avec listes

classification = ['Poids insuffisant','Poids normal','Excès de poids', 'Obésité, classe I', 'Obésité, classe II']
risque = ['Accru','Moindre','Accru','Elevé','Très élevé', 'Extrêmement élevé']
while True:
    try:
        poids = float(input('Saisir votre poid en kg '))
        if poids <= 0:
            print('entrer une valeur logique du poids')
        else:
            break
    except ValueError:
        print('entrer une valeur logique du poids')

while True:
    try:
        taille = float(input('Saisir votre taille en mètre '))
        if taille <= 0:
            print('entrer une valeur logique de la taille')
        else:
            break
    except ValueError:
        print('entrer une valeur logique de la taille')


imc = poids/taille**2
print(f'Votre IMC est {imc}')
if imc <18.5:
    print(f'Votre risque de développer des problèmes de santé est {risque[0]}')
    print(f'Votre classification est {classification[0]}')
elif imc <25:
    print(f'Votre risque de développer des problèmes de santé est {risque[1]}')
    print(f'Votre classification est {classification[1]}')
elif imc <30:
    print(f'Votre risque de développer des problèmes de santé est {risque[2]}')
    print(f'Votre classification est {classification[2]}')
elif imc <35:
    print(f'Votre risque de développer des problèmes de santé est {risque[3]}')
    print(f'Votre classification est {classification[3]}')
elif imc <40:
    print(f'Votre risque de développer des problèmes de santé est {risque[4]}')
    print(f'Votre classification est {classification[4]}')
else:
    print(f'Votre risque de développer des problèmes de santé est {risque[5]}')
    print(f'classification est {classification[5]}')
```
