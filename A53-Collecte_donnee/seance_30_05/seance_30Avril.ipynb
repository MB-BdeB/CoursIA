```python
# variable locale et globale 
def calculer_exposent(oper1:float,oper2:float=2)->float:
    '''
    fonction qui clacule la puissance d'un nombre
    '''
    resultat = oper1**oper2
    return resultat

#Appel de la fonction
resultat = calculer_exposent(5,3)
print(resultat)
```
